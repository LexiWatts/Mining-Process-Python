```python
# Data Visualizaiton
%pip install seaborn
import seaborn as sns
```

```python
# Data Visualization
%pip install matplotlib
import matplotlib.pyplot as plt
```

```python
# Read in the CSV data to pandas data storage
df = pd.read_csv('MiningProcess_Flotation_Plant_Database.csv',decimal=",")
```

```python
#Preview of Data using head function
df.head()

```

```python
#Determine number of rows and columns in the CSV
df.shape
```

```python
#Data preview of rows 100-104
df.iloc[100:105,:]
```

```python
#Testing to see the variable type imported
print(type(df))
print(type(df['date']))
print(type(df['date'][0]))
```

```python
#Redefine the date column to reflect as datetime column
df['date'] = pd.to_datetime(df['date'])
```

```python
#Descriptive analytics
df.describe()
```

```python
#Finding the max and min dates and times in the data set
max_date = df['date'].max()
print('The max date is ' + str(max_date))

min_date = df['date'].min()
print('The min date is ' + str(min_date))
```

```python
# Filter the data for just first week in june
df_june = df[(df['date'] > "2017-05-31 23:59:59") & (df['date'] < "2017-06-02")].reset_index(drop=True)
```

```python
#Data for first week in June
df_june 
```

```python
#Defined which columns are important and set data frame to June
important_cols = [
    'date',
    '% Iron Concentrate',
    '% Silica Concentrate',
    'Ore Pulp pH',
    'Flotation Column 05 Level'
]
#Setting new variable combining first week of june with the important columns
df_june_important = df_june[important_cols]
df_june_important
```

```python
#Call seaborn as sns and invoke the pairplot with the dataframe as an argument
sns.pairplot(df_june_important)
```

```python
#Determine if there is correlation between the previously defined variable
df_june_important.corr()
```

```python
#Created a line plot to see how the % Iron Concentrate changes throughout that day
sns.lineplot(x="date", y="% Iron Concentrate", data=df_june) 
```

```python
#To include other variables within the line plot I created a loop
for i in important_cols:
    sns.lineplot(x='date', y=i,  data=df_june)
    import matplotlib.pyplot as plt
    plt.show()
```

```python
#Created a histogram to view % Iron Concentrate during first week of June
sns.histplot(x="date", y="% Iron Concentrate", data=df_june) 
```

```python
# Add a column 'day_of_week' to the dataframe representing the day of the week
df['day_of_week'] = df['date'].dt.dayofweek

# Now plot the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='day_of_week', y='Ore Pulp pH', data=df)
plt.xlabel('Day of the Week (0=Monday, 6=Sunday)')
plt.ylabel('Ore Pulp pH')
plt.title('Distribution of Ore Pulp pH by Day of the Week')
plt.show()
```
